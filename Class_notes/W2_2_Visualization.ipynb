```python
sacramento.type.unique()
```

```python
sacramento ['sale_day_of_year'] = sacramento.clean_date.dt.dayofyear
```

```python
residential = sacramento[sacramento.type == 'Residential']
```

```python
condo = sacramento[sacramento.type == 'Condo']
```

```python
(fig, ax) = matplotlib.pyplot.subplots()
ax.plot(residential.groupby('sale_day_of_year').price.max(),
        c='green', label = 'Residential')
ax.plot(condo.groupby('sale_day_of_year').price.max(),
       c='magenta', label = 'Condo')
ax.set_xlabel("Day of the year")
ax.set_ylabel("Price in ($)")
ax.legend(loc ='right')
```

```python
(fig, ax) = matplotlib.pyplot.subplots()
ax.plot(residential.groupby('sale_day_of_year').price.max(),
        c='green', label = 'Residential', marker = "o")
ax.plot(condo.groupby('sale_day_of_year').price.max(),
       c='magenta', label = 'Condo',
       marker = '^', linestyle = "--")
ax.set_xlabel("Day of the year")
ax.set_ylabel("Price in ($)")
ax.legend(loc ='right')
```
